# Quick guide

#### 1. Upload the MD trajectories and the reference structure into MDAnalysis Universes <span style='color:darkorange'> ***u***  </span> and <span style='color:darkorange'> ***u_ref***  </span>repectively
#### 2. Select NbIT sites and create MDAnalysis selection objects for them or/and assemble them into the list (COMB).
#### 3. Create NbIT object: *NbIT(u, u_ref, mode)*
- <span style='color:hotpink'> ***u***  </span> - MDAnalysis Universe with MD trajectory
- <span style='color:hotpink'> ***u_ref***  </span> - MDAnalysis Universe with reference structure
- <span style='color:hotpink'> ***mode***  </span> - mode (atoms/residues) considered for all NbIT functions except total_intercorrelation (*default*: residues) (see [1])
#### 4. Evaluate the covariance matrix (COV) by calling *compute_cov(selection, window, overlap)* function. If sliding window regime was selected the resulting array would contain several covariance matrices  - one for each window (see also function comments) . Function parameters:
- <span style='color:hotpink'> ***selection*** </span> (str) - selection line of the atoms for COV calculation. *Example*: for heavy atoms selection='not type H*'
- <span style='color:hotpink'> ***window*** </span> - size of window for COV evaluation. If window = nframes the entire trajectory will be considered
- <span style='color:hotpink'> ***overlap*** </span> - the size of overlap for slading window. *Default*: 0

#### 5. NbIT functions are called as standard python class functions. 
#### ***Example:*** 
    O = NbIT(u, u_ref)   # created NbIT object
    O.compute_cov(selection = 'not type H*', window = 1000, overlap = 100)   # compute COV for heavy atoms
    O.coordination(COMB)    # calculate the Coordination Information values for NbIT sites in COMB list
    
#### 6. Some of the NbIT functions available (see also comments to functions):

- **mutual(COMB)** - mutual information for NbIT sites in COMB list
- **coordination(COMB)** - normalized coordiantion information for NbIT sites in COMB list
- **contribution(COMB[i])** - individual contribution of each residue into the total correlation of NbIT site *i*
- **average_contribution_coordination(COMB[i], COMB[j])** - individual contribution of each residue into coordination information between *i* and *j* sites
- **mutual_channel(COMB[i], COMB[j])** - co-information shared between *i* and *j* sites and each residue of the system
- **coordination_channel(COMB[i], COMB[j])** - coordination information shared between sites *i* and *j* and each residues of the system
- **normalized_coordination_channel(COMB[i], COMB[j])** - normalized values of coordination channel
- **total_intercorrelation_coeff(COMB[i], COMB[j], mode_inter)** - total intercorrelation coeeficient; mode_inter(*default*) = 'atoms' (see [2])

>[1] LeVine M.V., Weinstein H. (2014) NbIT - A New Information Theory-Based Analysis of Allosteric Mechanisms Reveals Residues that Underlie Function in the Leucine Transporter LeuT. PLoS Comput Biol 10(5): e1003603. doi:10.1371/journal.pcbi.1003603
<br>
>[2] Khelashvili, G.; et al. Spontaneous Inward Opening of the Dopamine Transporter Is Triggered by PIP2‑Regulated Dynamics of the N‑Terminus. 2015, ACS Chem Neurosci., 6, 1825-1837.

```python
import MDAnalysis as mda
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import log, exp, pi
```

```python
class NbIT:

    
    def __init__(self, u, u_ref, mode = 'residues'):
        self.u = u
        self.u_ref = u_ref
        self.u_ref.add_TopologyAttr('tempfactors')
        self.mode = mode
        

    def compute_cov(self, selection, window, overlap): 
        
    ####################
    #    COV matrix    #
    ####################

    ### Parameters
    #
    ### window    - Size of a window for COV evaluation. If nsteps = nframes (number 
    #               of frames in the trajectory) the COV will be evaluated from the 
    #               entire trajectory. If however nsteps < nframes, the COV will be
    #               evaluated for the trajectory segments of nsteps size. The output in 
    #               this case will contain nframes//nsteps number of COV matrices (for 
    #               overlap = 0).
    #
    ### overlap   - Overlap window size for COV matrix calculation. If overlap != 0
    #               and nsteps < nframes, each consecutive COV will include the specified
    #               number of frames from the preceeding window. 
    #
    #               Example:
    #               for a trajectory with 654 frames window = 100 and overlap = 10 the 
    #               segments would be [1-100, 90-190, 180-280, 270-370, 360-460, 
    #               450-550, 540-640] 
    ###


        self.selection = selection
        self.window = window
        self.overlap = overlap
        self.atoms = self.u.select_atoms(selection)
        self.atoms_ref = self.u_ref.select_atoms(selection)

        # Get the atom positions in reference structure
        p0 = self.atoms_ref.positions

        # Compute the distance matrices
        N = (self.u.trajectory.n_frames-self.window)//(self.window - self.overlap)
        if self.u.trajectory.n_frames >= (N+1)*self.window - self.overlap*N:
            N +=1
        dist_matrix = np.empty((p0.shape[0], self.window, N))
        for n in range(N):
            for ts in self.u.trajectory[(self.window-self.overlap)*n : self.window*(n+1) - self.overlap*n]:
                dist_matrix[:, ts.frame - ((self.window - self.overlap)*n), n] = np.linalg.norm(self.atoms.positions-p0, axis = -1)[:]


        # Compute covariance matrix
        cov = np.empty((p0.shape[0], p0.shape[0], N))
        for n in range(N):
            cov [:, :, n] = np.cov(dist_matrix[:, :, n], rowvar=True)
        self.COV = cov
        (self.L, self.L, self.N) = self.COV.shape
        
    def cov_save(self, filename):
        
        self.filename = filename
        np.save(self.filename, self.COV)
        
    def cov_load(self, filename, selection):
        
        self.filename = filename
        self.selection = selection
        self.atoms = self.u.select_atoms(selection)
        self.atoms_ref = self.u_ref.select_atoms(selection)        
        
        self.COV = np.load(self.filename)
        (self.L, self.L, self.N) = self.COV.shape
                
    
    def entropy(self, S):
        H = np.empty(self.N)
        for n in range(self.N):
            j = np.where(np.in1d(self.atoms_ref.ix, np.unique(S.ix))==True)[0]                        
            if j.size>1:
                H[n] = 0.5*j.size*(1+log(2*pi))+0.5*np.linalg.slogdet(self.COV[:, :, n].reshape(self.L, -1)[:, j][j, :])[1]
            elif j.size ==1:
                H[n] = 0.5*(log(2*pi*exp(1)*self.COV[:, :, n][:, j][j]))
            
            else:
                return print('Wrong index format or no elements in the list were found')

        return H


    def inf(self, S1, S2):
        
        if S1.ids.size>=1 and S2.ids.size>=1:
            H1 = self.entropy(S1)
            H2 = self.entropy(S2)
            H12 = self.entropy(S1+S2)
            return abs(H1 + H2 - H12)
        else:
            print('Wrong index format or no elements in the list were found')

    def mutual(self, COM):

    ########################################
    # If any two NbIT sites contain mutual residue it would be attributed to the earliest site 
    # difined in COM
    ########################################
    
        length = len(COM)
        mutual_out = np.empty((length, length, self.N))
        for i in range(length):
            for j in range(length):
                if i==j:
                    mutual_out[i, j, :] = self.entropy(COM[i])
                elif i< j:
                    S1 = COM[i]
                    S2 = COM[j].difference(COM[i])
                    mutual_out[i, j, :] = self.inf(S1, S2)
                else:
                    mutual_out[i, j, :] = 'NaN'
        return mutual_out


    def total_correlation(self, S, mode):
        
    #############################
    # Total correlation in the majority of the functions is supposed to be calculated per-residue wise according to [1]
    # However, to calcualte total_intercorrelation and total_intercorrelation_coeff the per-atom version is used [2]
    # For total_intercorrelation related calculations the per-atom mode is selected automaticly, however it could be used for
    # any NbIT function calculation

    # [1] LeVine M.V., Weinstein H. (2014) NbIT - A New Information Theory-Based Analysis of Allosteric Mechanisms Reveals Residues that Underlie Function in the
    # Leucine Transporter LeuT. PLoS Comput Biol 10(5): e1003603. doi:10.1371/journal.pcbi.1003603
    #
    # [2] Khelashvili, G.; et al. Spontaneous Inward Opening of the Dopamine Transporter Is Triggered by PIP2‑Regulated 
    # Dynamics of the N‑Terminus. 2015, ACS Chem Neurosci., 6, 1825-1837.
    #############################

        sum_H = 0
        if mode == 'residues':
            for r in np.unique(S.resindices):
                sum_H += self.entropy(S[np.where(S.resindices == r)[0]])
        else:
            for a in S.atoms:
                sum_H += self.entropy(a)
        H = self.entropy(S)
        return sum_H - H

    def conditional_total_correlation(self, S1, S2, mode):
        
        sum_H = 0
        if mode == 'residues':
            for r in np.unique(S1.resindices):
                sum_H += self.conditional_entropy(S1[np.where(S1.resindices == r)[0]], S2)
        else:
            for a in np.unique(S1.ids):
                sum_H += self.conditional_entropy(S1[np.where(S1.ids == a)[0]], S2)
        H = self.conditional_entropy(S1, S2)
        return sum_H - H    
    
    def total_intercorrelation(self, S1, S2, mode_inter = 'atoms'):
        return self.total_correlation((S1+S2).unique, mode_inter) - self.conditional_total_correlation(S1, S2, mode_inter) - self.conditional_total_correlation(S2, S1, mode_inter)

    def total_intercorrelation_coeff(self, S1, S2, mode_inter = 'atoms'):
        
        if mode_inter =='residues':
            scaling_coeff = np.unique(S1.unique.resindices).size + np.unique(S2.unique.resindices).size - 1 
        else:
            scaling_coeff = S1.unique.ids.size + S2.unique.ids.size - 1             
        exp_part = np.exp(-2*self.total_intercorrelation(S1, S2, mode_inter)/scaling_coeff)
        return np.sqrt(1 - exp_part)

    def conditional_entropy(self, S1, S2):

    ##########################
    # If any two NbIT sites contain mutual residue it would be attributed to the site 
    # on which we condition in COM
    ##########################

        H12 = self.entropy((S1+S2).unique)
        H2 = self.entropy(S2)
        return H12 - H2

    def coordination_information(self, S1, S2):
        return (self.total_correlation(S1, mode = self.mode) - self.conditional_total_correlation(S1, S2, mode = self.mode))

    def normalized_cooordination_information(self, S1, S2):
        return self.coordination_information(S1, S2)/self.total_correlation(S1, mode = self.mode)

    ## COORDINATION

    def coordination(self, COM):

    ##########################
    # If any two NbIT sites contain mutual residue it would be attributed to the earliest site 
    # difined in COM
    ##########################
    
        length = len(COM)
        coordination_output = np.ones((length, length, self.N))
        for i in range(length):
            for j in range(length):
                if i == j:
                    coordination_output[i, j, :] = self.total_correlation(COM[i], mode = self.mode)
                else:
                    s1 = COM[i]
                    s2 = COM[j].difference(COM[i])
                    if s1.ids.size*s2.ids.size==0:
                        coordination_output[i, j, :] = 'NaN'
                    else:
                        coordination_output[i, j, :] = self.normalized_cooordination_information(s1, s2) 
        return coordination_output

    ## Individual contribution

    def contribution(self, S1):
        
        l = np.unique(S1.resindices)
        all_contrib_output = np.empty((l.size, self.N))
        for r in range(l.size):
            S2 = S1[np.where(S1.resindices == l[r])[0]]
            all_contrib_output[r, :] = 1-self.conditional_total_correlation(S1, S2, mode = self.mode)/self.total_correlation(S1, mode = self.mode)
        return all_contrib_output


    def conditional_information(self, S1, S2, r):
        
        s1 = S1.difference(r)
        s2 = S2.difference(r)
        H1 = self.conditional_entropy(s1, r)
        H2 = self.conditional_entropy(s2, r)
        H12 = self.conditional_entropy((s1+s2).unique, r)
        return H1 + H2 - H12


    ## Mutual contribution

    def average_contribution_mutual(self, S1, S2):
        S12 = (S1+S2).unique
        l = np.unique(S12.resindices)
        all_contrib_mutual = np.empty((l.size, self.N))
        for r in range(l.size):
            res = S12[np.where(S12.resindices == l[r])[0]] 
            all_contrib_mutual[r, :] = 1-self.conditional_information(S1, S2, res)/self.inf(S1, S2)
        return all_contrib_mutual

    def conditional_coordination_information(self, S1, S2, r):
        return self.conditional_total_correlation(S1, r, mode = self.mode) - self.conditional_total_correlation(S1, S2+r, mode = self.mode)

    ## Contribution coordination

    def average_contribution_coordination(self, S1, S2, ax = None, **plt_kwargs):
        
        S12 = (S1+S2).unique
        l = np.unique(S12.resindices)
        all_contrib_coord = np.empty((l.size, self.N))
        for r in range(l.size):
            res = S12[np.where(S12.resindices == l[r])[0]]
            all_contrib_coord[r, :] = 1 - self.conditional_coordination_information(S1, S2, res)/self.coordination_information(S1, S2)
        if ax is None:
            fig, ax = plt.subplots(figsize = (10, 8))
        resnames_1 = [str(k).replace('>', '').replace('<', '').replace(',', '').split(maxsplit=1)[1] for k in S1.residues]
        resnames_2 = [str(k).replace('>', '').replace('<', '').replace(',', '').split(maxsplit=1)[1] for k in S2.residues]
        
        # Account for mutual residues in COMB[i] and COMB[j] and assign them to the first site 
        # (following the convention of average contribution coordination function)
        
        inter = list(set(resnames_1).intersection(resnames_2))
        if len(inter) !=0:
            for res in inter:
                resnames_2.remove(res)

        ax.plot(all_contrib_coord, **plt_kwargs)
        ax.set_xticks(np.arange(all_contrib_coord.shape[0]))
        ax.set_xticklabels(resnames_1+resnames_2, size =12, rotation = 45)
        ax.axvline(x = (len(resnames_1)-0.5), ymin = 0, ymax = 1, ls = '--', color = 'black', label = 'site division')
        ax.legend()
        return all_contrib_coord

    def co_information(self, S1, S2, r):
        
        part = self.conditional_information(S1, S2, r)
        total = self.inf(S1, S2)
        return total - part

    ## Mutual Channel

    def mutual_channel(self, S1, S2):

    ##########################
    # Co-information values for the residues in S1 or S2
    # are set to 0
    ##########################
    
        l = np.unique(self.atoms.resindices)
        score = np.empty((l.size, self.N))
        for r in range(l.size): 
            score[r, :] = self.co_information(S1, S2, self.atoms[np.where(self.atoms.resindices == l[r])[0]])
        for r in np.unique((S1+S2).unique.resindices):
            score[r-1, :] = 0
        return score

    def mutual_coordination_information(self, S1, S2, r):
        return self.coordination_information(S1, S2) - self.conditional_coordination_information(S1, S2, r)

    ## Coordination channel

    def coordination_channel(self, S1, S2, save_to_pdb = False, pdbfile_name = None):
        
        l = np.unique(self.atoms.resindices)
        score = np.empty((l.size,self.N))
        for r in range(l.size):
            score[r, :] = self.mutual_coordination_information(S1, S2, self.atoms[np.where(self.atoms.resindices == l[r])[0]])
        if save_to_pdb:
            betas = np.empty(self.u_ref.trajectory.n_atoms)
            for i in range(score.shape[0]):
                atom1 = int(str(self.u_ref.residues[i].atoms[0]).split(':')[0].split(' ')[1])
                atom_1 = int(str(self.u_ref.residues[i].atoms[-1]).split(':')[0].split(' ')[1])
                betas[atom1-1:atom_1] =score[i]
            betas = np.round_(betas, 2)

            try:
                pdb_out = pdbfile_name
                with mda.Writer(pdb_out, multiframe=False, bonds=None, n_atoms=self.N) as PDB:
                    self.u_ref.atoms.tempfactors = betas
                    PDB.write(self.u_ref.atoms)                
            except:
                print('pdbfile_name is required when save_to_pdb is on')

        return score
        

    def normalized_mutual_coordination_information(self, S1, S2, r):
        
        return self.mutual_coordination_information(S1, S2, r)/self.coordination_information(S1, S2)

    def normalized_coordination_channel(self, S1, S2, save_to_pdb = False, pdbfile_name = None):
        
        l = np.unique(self.atoms.resindices)
        score = np.empty((l.size, self.N))
        for r in range(l.size):
            score[r, :] = self.normalized_mutual_coordination_information(S1, S2, self.atoms[np.where(self.atoms.resindices == l[r])[0]])
        if save_to_pdb:
            betas = np.empty(self.u_ref.trajectory.n_atoms)
            for i in range(score.shape[0]):
                atom1 = int(str(self.u_ref.residues[i].atoms[0]).split(':')[0].split(' ')[1])
                atom_1 = int(str(self.u_ref.residues[i].atoms[-1]).split(':')[0].split(' ')[1])
                betas[atom1-1:atom_1] =score[i]
            betas = np.round_(betas, 2)
            try:
                pdb_out = pdbfile_name
                with mda.Writer(pdb_out, multiframe=False, bonds=None, n_atoms=self.N) as PDB:
                    self.u_ref.atoms.tempfactors = betas
                    PDB.write(self.u_ref.atoms)                
            except:
                print('pdbfile_name is required when save_to_pdb is switched on')

        return score
```

```python
## Pandas representation

def high_diff(s):
    return ['background-color: pink' if v> 0.10 else 'background-color: lightblue' if v<-0.10 else '' for v in s]
def high(s):
    return ['background-color: pink' if v > 0.2 else '' for v in s]
```

```python
# Load the trajectory assuming it is already aligned (see align.ipynb)
u = mda.Universe('./cut.psf',
                 './all.xtc')
u_ref = mda.Universe('./cut.psf',
                     './cut.pdb')
```

```python
# Select the regions for NbIT
S1 = u.select_atoms('(segid LIG and resid 519) or (segid PROT and resid 25 26 104 108 253 254 256 259 355 359) or (segid INT and resid 517) and not type H')
NA1 = u.select_atoms('(segid INT and resid 517) or (segid PROT and resid 22 27 254 286) and not type H')
NA1 = u.select_atoms('(segid INT and resid 517) or (segid PROT and resid 22 27 254 286) or (segid LIG and resid 519) and not type H')
NA2 = u.select_atoms('(segid INT and resid 516) or (segid PROT and resid 20 23 351 354 355) and not type H')
S2 = u.select_atoms('(segid PROT and resid 29 30 107 111 114 253 319 320 324 400 404) and not type H')
INI = u.select_atoms('(segid PROT and resid 5 187 267 268 361 369) and not type H')
```

```python
# Select combinations of regions for the analysis
COMB = [S1, S2, NA1, NA2, NA1+NA2, NA1+NA2+S1, NA1+NA2+S1+S2, INI]
names = ['S1', 'S2', 'NA1', 'NA2', 'NA1+NA2', 'NA1+NA2+S1', 'NA1+NA2+S1+S2', 'INI']
```

```python
O = NbIT(u, u_ref)
O.compute_cov(selection = 'not type H*', window = u.trajectory.n_frames, overlap = 0)
```
